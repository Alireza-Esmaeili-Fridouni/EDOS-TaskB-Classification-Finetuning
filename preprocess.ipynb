```python
import pandas as pd 
import re
```

```python
data = pd.read_csv(f"data\edos_labelled_aggregated.csv")
data
```

```python
data = data[["text", "label_category", "split"]]
```

```python
data.duplicated().sum().item()
```

```python
data['label_category'].value_counts(sort=True, ascending=True)
```

```python
data.loc[:,'label_type'] = data['label_category'].str.replace(r"^\d+\.\s*", "", regex=True)
data.loc[:,"category_code"] = pd.to_numeric(data["label_category"].str.extract(r"^(\d+)\.")[0], errors='coerce').fillna(4).astype(int)
data
```

```python
splits = {split_name: group for split_name, group in data.groupby('split')}
```

```python
train_df = splits.get('train')
test_df = splits.get('test')
val_df = splits.get('dev')
```

```python
train_df.drop(['split'], axis=1, inplace=True)
test_df.drop(['split'], axis=1, inplace=True)
val_df.drop(['split'], axis=1, inplace=True)
```
